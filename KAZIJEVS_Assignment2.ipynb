### Question 2
Write a python code to draw the given line equation at the below:
f(x)= x^7 - 8x^3+ 5x  between -2, 1. 
Use annotating text property to show the maximum of this function.


Your Answer:

```python
def f(x):
    res = (x**7) - (8*(x**3)) + (5*x)
    return res

x = np.arange(-2.0, 1, 0.05)

plt.figure(figsize=(12,6))
plt.plot(x, f(x))
plt.annotate("Local max ({:.3f}, {:.3f})".format(x[np.argmax(f(x))], max(f(x)) ), xy=(x[np.argmax(f(x))], max(f(x))), xytext=(-1, -30),
arrowprops=dict(facecolor='black', shrink=0.01))
plt.show()
```

### Question 3
#### Graduate Student Question / BONUS Question for Undergraduate Students
The daily profit, P, of an oil refinery is given by

P = 8x − 0.005x^2,
	    
where x is the number of barrels of oil refined. How many barrels will give maximum profit and what is the maximum profit? Write a code that creates 10000 points (x values) and find the x value for the maximum profit. Plot all these points and show your results.


```python
def p(x):
    return (8*x) - (0.005*x**2)

x = np.arange(1, 10001, 1)
y = p(x)
maxP = max(y)
barrelMax = x[np.argmax(y)]
print("Max profit")
print(maxP)
print("Number of barrels that give max profit")
print(barrelMax)

plt.figure(figsize=(12,6))
plt.plot(x, y, 'g^')
plt.xlabel('Number of barrels')
plt.ylabel('Profit')
plt.annotate('Max Profit x = {} y = {}'.format(barrelMax, maxP), xy=(barrelMax, maxP),
xytext=(barrelMax, maxP-100000),
             arrowprops=dict(facecolor='black', shrink=0.05)) 
plt.show()
```
