### Question 1
Write your OWN Bubble Sort code in python to sort these numbers: <br>
43, 5, 7, 12, 100, 47, 12, 15, 33, 8, 67, 14, 98, 86 <br>
Your Answer:


```python
list_arr = [43, 5, 7, 12, 100, 47, 12, 15, 33, 8, 67, 14, 98, 86]            
```

```python
def bubbleSort(arr):
    for i in range(len(arr)-1):
        for j in range(len(arr)-1):
            if arr[j] > arr[j+1]:
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
                
    return arr
```

```python
bubbleSort(list_arr)
list_arr
```

### Question 2
Use the code and the answer from the first question to write your OWN Binary Search code to search for a given number.<br>
Your Answer:

```python
def binarySearch(arr, target):
    left = 0
    right = len(arr)-1

    while left <= right:
        middle = int((left+right)/2)
        if target > arr[middle]:
            left = middle+1
        if target < arr[middle]:
            right = middle-1
        if target == arr[middle]:
            return middle

    return -1
```

```python
number = 67 # Enter your number here
index = binarySearch(list_arr, number)

if index == -1:
    print("Number was not found in the list")
else:
    print("Number", number, "was found at index:",index)
```

Dynamic Programming is an algorithm design technique for optimization problems. Like the divide-and-conquer method, dynamic programming solves problems by combining the solutions to sub-problems.

Write your OWN dynamic programming version of recursive Fibonacci code using the pseudo code given below:


Keep in mind that f is the array you store the sequence. This f shall be defined global and set all elements to -1.
Use your code to show the first 10 numbers in the series. 


```python
f = np.full(100,-1)
```

```python
def myfib2(n):
    global f
    if n == 0:
        f[0] = 0
        return f[0]
    if n == 1:
        f[1] = 1
        return f[1]
    if f[n-2] == -1:
        f[n-2] = myfib2(n-2)
    if f[n-1] == - 1:
        f[n-1] = myfib2(n-1)
        
    f[n] = f[n-1] + f[n-2]
    return f[n]
```
